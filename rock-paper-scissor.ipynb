```python
import random

player = int(input("What do you choose? Type 0 for Rock, 1 for Paper and 2 for Scissor\n"))
computer = random.randint(0,2)

def check(x, y):
    if x == y:
        print("You are tied")
    elif x == 0 and y == 2:
        print("You win")
    elif x == 2 and y == 0:
        print("You lose")
    elif x > y:
        print("You win")
    elif y > x:
        print("You lose")        
    else:
        print("Invalid")

if player == 0:
    print("""
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
""")
elif player == 1:
    print("""
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
""")
elif player == 2:
    print("""
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
""")

if computer == 0:
    print("Computer chose:\n" + """
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)
""")
elif computer == 1:
    print("Computer chose:\n" + """
     _______
---'    ____)____
           ______)
          _______)
         _______)
---.__________)
""")
elif computer == 2:
    print("Computer chose:\n" + """
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)
""")
check(player, computer)
    
    
```
