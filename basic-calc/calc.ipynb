```python
import art
from IPython.display import clear_output

print(r"{}".format(art.logo))

def calculate(first: int, second: int, ope: str):
    """
    Calculate the first and second numerical input based on the operator the user has chosen. e.g x + y. 
    It then determine if it will either end or continue the calculation of the output or two new inputs.
    
    Args:
        first (int): The first input as x
        second (int): The second input as y
        ope (str): The type of math operator between '+,-,*,/'
    """
    x = first
    y = second
    operator = {
    "+": x + y,
    "-": x - y,
    "*": x * y,
    "/": x / y 
    }
    ope_check = operator.get(ope, "Invalid Operator")
    print(ope_check)
    if ope_check == "Invalid Operator":
        clear_output(wait=True)
        print("There was error input in the calculation")
        return
    solution = operator[ope]
    print(f"{first}{ope}{second} = {solution}")
    cont = input(f"Type 'again' to continue calculating with {solution}, type 'new' to start a new calculation, or 'end' to end calculation: ")
    if cont == 'again' and not ope_check == "Invalid Operator":
        new_ope = input("Pick an operator: ")
        new_s = input("What's the next number?: ")
        if new_s.isdigit():
            calculate(int(solution),int(new_s),new_ope)
        else:
            while not new_s.isdigit():
                new_s = input("What's the next number?: ")
            calculate(solution,new_s,new_ope)
    elif cont == 'new' and not ope_check == "Invalid Operator":
        clear_output(wait=True)
        compute()
    else:
        return

def compute():
    """
    The controller for the calculate(num1,num2,str) 
    """
    f_input = input(" \nWhat's the first number?: ")
    print(" + \n - \n * \n /")
    operator = input("Pick an operator: ")
    s_input = input("What's the next number?: ")
    if f_input.isdigit() and s_input.isdigit():
        calculate(int(f_input), int(s_input), operator)
    else:
        if f_input == "" or operator == "" or s_input == "":
            return
        else:
            while not f_input.isdigit() and not s_input.isdigit():
                clear_output(wait=True)
                compute()

compute()
```
